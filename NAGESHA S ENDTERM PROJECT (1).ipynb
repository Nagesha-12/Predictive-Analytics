#Linear regression analysis is used to predict the value of a variable based on the value of another variable. The variable you want to predict is called the dependent variable. The variable you are using to predict the other variable's value is called the independent variable.

```python
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
np.corrcoef(data.RunsScored,data.ValueinCR)[0][1]
```

The Correlation Co-efficient  0.14 is an +ve Correlation. There is a relation Between Both the Varialbles

```python
plt.scatter(data.RunsScored,data.ValueinCR)
plt.xlabel('RunsScored')
plt.ylabel('ValueinCR')
```

```python
#Store the features in X and target in y
```

```python
X=data[['RunsScored']]
y=data[['ValueinCR']]
```

Here Y is an independent variable and X is dependent variable because if the score more runs they will get paid more.

```python
#Split the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

 the sklearn model will select the data to train and test by spiliting the data. the test data is 20% then the train data is 80%. 

```python
#hstack stacks two column arrys horizontally - side by side
pd.DataFrame(np.hstack((X_train,y_train)),columns=['Xtrain','ytrain']).head()
```

```python
print(X_train.shape,y_test.shape)
```

here the 1 is represents the independent variable.

```python
#Create an instance of Linear regression
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
```

```python
#Train the model
regr.fit(X_train,y_train)
```

```python
#Get the slope and intercept
print("a1 = ", regr.coef_)
print("a0 = ",regr.intercept_)
```

y=a=bx here the a0 is a its intercept, a1 is an b its coef.

```python
#Test the model : Make predictions on the trained model
y_pred = regr.predict(X_test)
```

regr is an code for regression. predicting the X_test

```python
y_pred.reshape(-1,1)
```

```python
#Reshape here converts a 1d vector into a 2d column vector
pd.DataFrame(np.hstack((X_test,y_test,y_pred.reshape(-1,1))),columns=['x_test','y_test','y_pred'])
```

```python
round(np.sqrt(mean_squared_error(y_test,y_pred)),4)
```

the mean squared error is say the either it can be + or - to the prediction that is 42142530.5399 it can be added or subracted.

```python
round(r2_score(y_test,y_pred),4)
```

if the variation in y can be expalined 62% accuarately variation with x. but the negative r square is not considered.

```python
plt.scatter(X_test,y_test,color = 'red')
plt.plot(X_test,y_pred,color='blue')
plt.title("RunsScored v/s ValueinCR(Test set)")
plt.xlabel('RunsScored')
plt.ylabel('ValueinCR')
plt.show()
```
